````python
import os
import requests
from github import Github, GithubException
import traceback
import re

def test_submit_comment(access_token, repo_name, pr_number, path, line_number, body):
    g = Github(access_token)
    
    try:
        user = g.get_user()
        print(f"Authenticated as: {user.login}")
    except GithubException as e:
        print(f"Authentication failed: {e.status} - {e.data.get('message', '')}")
        return
    
    try:
        repo = g.get_repo(repo_name)
        print(f"Successfully accessed repo: {repo.full_name}")
    except GithubException as e:
        print(f"Error accessing repo: {e.status} - {e.data.get('message', '')}")
        return
    
    try:
        pr = repo.get_pull(pr_number)
        print(f"Successfully accessed PR #{pr.number}")
        print(f"PR Title: {pr.title}")
        print(f"PR State: {pr.state}")
    except GithubException as e:
        print(f"Error accessing PR: {e.status} - {e.data.get('message', '')}")
        return
    
    commit_sha = pr.head.sha
    print(f"Head commit SHA: {commit_sha}")
    
    try:
        files = pr.get_files()
        print(f"PR has {files.totalCount} files changed.")
    except GithubException as e:
        print(f"Error fetching PR files: {e.status} - {e.data.get('message', '')}")
        return
    
    for file in files:
        if file.filename == path:
            print(f"Found file: {path} with {file.changes} changes.")
            patch = file.patch
            if not patch:
                print(f"No patch available for {path}. Cannot add a review comment.")
                return
            
            print(f"Patch content:\n{patch}")
            
            patch_lines = patch.split('\n')
            
            position = None
            diff_hunk = None
            current_line = 0
            for i, diff_line in enumerate(patch_lines):
                if diff_line.startswith('@@'):
                    match = re.search(r'@@ -\d+(?:,\d+)? \+(\d+)(?:,\d+)? @@', diff_line)
                    if match:
                        current_line = int(match.group(1))
                    else:
                        print(f"Failed to parse diff header: {diff_line}")
                    continue
                
                if diff_line.startswith('+') and not diff_line.startswith('+++'):
                    if current_line == line_number:
                        position = i
                        # Get the diff hunk (context)
                        start = max(0, i - 3)
                        end = min(len(patch_lines), i + 4)
                        diff_hunk = '\n'.join(patch_lines[start:end])
                        break
                    current_line += 1
                elif not diff_line.startswith('-'):
                    current_line += 1
            
            if position is None:
                print(f"Line {line_number} is not part of the diff and cannot be commented on.")
                print(f"The diff affects lines {current_line - file.changes} to {current_line - 1}.")
                return
            
            print(f"Commenting on position {position} (line {line_number} in the new file).")
            break
    else:
        print(f"File {path} not found in the PR.")
        return
    
    # Use a low-level API call to submit the comment
    url = f"https://api.github.com/repos/{repo_name}/pulls/{pr_number}/comments"
    headers = {
        "Authorization": f"token {access_token}",
        "Accept": "application/vnd.github.v3+json"
    }
    data = {
        "body": body,
        "commit_id": commit_sha,
        "path": path,
        "position": position,
        "line": line_number,
        "side": "RIGHT",
        "diff_hunk": diff_hunk
    }
    
    response = requests.post(url, headers=headers, json=data)
    
    if response.status_code == 201:
        print("Comment submitted successfully.")
        print(f"Comment URL: {response.json()['html_url']}")
    else:
        print(f"Error submitting comment. Status code: {response.status_code}")
        print(f"Response content: {response.content}")

# Usage
access_token = os.getenv('GITHUB_ACCESS_TOKEN')
repo_name = "isavita/sudoku"
pr_number = 1
path = "README.md"
line_number = 21  # This should be a line number that's part of the diff
body = "This is a test comment. Please review this change carefully."

test_submit_comment(access_token, repo_name, pr_number, path, line_number, body)
````

````python
import re
import requests
from github import GithubException

def submit_line_comment(repo, pr_number, path, line, body, commit_id):
    pr = repo.get_pull(pr_number)
    url = f"https://api.github.com/repos/{repo.full_name}/pulls/{pr_number}/comments"
    headers = {
        "Authorization": f"token {repo._Github__requester._Requester__authorizationHeader.split(' ')[1]}",
        "Accept": "application/vnd.github.v3+json"
    }
    data = {
        "body": body,
        "commit_id": commit_id,
        "path": path,
        "line": line,
        "side": "RIGHT"
    }
    
    response = requests.post(url, headers=headers, json=data)
    
    if response.status_code == 201:
        return response.json()
    else:
        return None

def get_change_line_numbers(patch):
    lines = patch.split('\n')
    header = lines[0]
    
    match = re.search(r'@@ -\d+(?:,\d+)? \+(\d+)(?:,\d+)? @@', header)
    if not match:
        raise ValueError("Unable to parse patch header")
    
    new_start = int(match.group(1))
    additions = sum(1 for line in lines[1:] if line.startswith('+'))
    end_line = new_start + additions - 1
    
    return new_start, end_line

def comment_on_change(repo, pr_number, file_change, comment_text, position='top'):
    path = file_change['filename']
    patch = file_change['patch']
    
    try:
        start_line, end_line = get_change_line_numbers(patch)
    except ValueError:
        return None
    
    line = start_line if position == 'top' else end_line
    
    commit_id = repo.get_pull(pr_number).head.sha
    
    return submit_line_comment(repo, pr_number, path, line, comment_text, commit_id)

repo = "isavita/sudoku"
pr_number = 1
file_changes = get_pr_file_changes(repo, pr_number)
for file_change in file_changes:
    comment_text = "Please review this change carefully [top]"
    result = comment_on_change(repo, pr_number, file_change, comment_text, position='top')
    
    comment_text = "Please review this change carefully [bottom]"
    result = comment_on_change(repo, pr_number, file_change, comment_text, position='bottom')
    if result:
        print(f"Comment added successfully. URL: {result['html_url']}")
    else:
        print("Failed to add comment.")
````
