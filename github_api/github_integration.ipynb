````python
import os
import re
from github import Github, GithubException

class GitHubIntegration:
    def __init__(self, access_token):
        self.github = Github(access_token)

    def get_repo(self, repo_name):
        try:
            return self.github.get_repo(repo_name)
        except GithubException as e:
            if e.status == 404:
                return None
            else:
                raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_pull_request(self, repo, pr_number):
        try:
            return repo.get_pull(pr_number)
        except GithubException as e:
            if e.status == 404:
                return None
            else:
                raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_pull_requests(self, repo, state='open'):
        return list(repo.get_pulls(state=state))

    def get_pr_files(self, pull_request):
        try:
            return list(pull_request.get_files())
        except GithubException as e:
            raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_file_patch(self, file):
        return file.patch

    def parse_patch(self, patch):
        if not patch:
            return None  # Return None for empty patches

        lines = patch.split('\n')
        header = lines[0]
        changes = lines[1:]
        
        # Parse the header
        match = re.match(r'@@ -(\d+)(?:,(\d+))? \+(\d+)(?:,(\d+))? @@', header)
        if not match:
            raise ValueError("Invalid patch header format")
        
        old_start = int(match.group(1))
        old_lines = int(match.group(2) or 1)  # Default to 1 if not specified
        new_start = int(match.group(3))
        new_lines = int(match.group(4) or 1)  # Default to 1 if not specified
        
        additions = [line[1:] for line in changes if line.startswith('+')]
        deletions = [line[1:] for line in changes if line.startswith('-')]
        context = [line[1:] for line in changes if line.startswith(' ')]
        
        is_new_file = old_start == 0 and old_lines == 0
        is_deleted_file = new_start == 0 and new_lines == 0
        
        if is_new_file:
            end_line = new_start + len(additions) - 1
        elif is_deleted_file:
            end_line = old_start + len(deletions) - 1
        else:
            end_line = new_start + len(additions) - len(deletions) + old_lines - 1
        
        return {
            'old_start': old_start,
            'old_lines': old_lines,
            'new_start': new_start,
            'new_lines': new_lines,
            'additions': additions,
            'deletions': deletions,
            'context': context,
            'start_position': 1,  # GitHub API uses 1-based indexing for positions
            'end_position': len(changes),
            'end_line': max(1, end_line),  # Ensure end_line is at least 1
            'is_new_file': is_new_file,
            'is_deleted_file': is_deleted_file
        }

    def create_comment(self, pull_request, body, commit, path, position):
        try:
            comment = pull_request.create_review_comment(body, commit, path, position)
            return comment
        except GithubException as e:
            raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")


def run_tests():
    token = os.getenv("GITHUB_ACCESS_TOKEN")
    if not token:
        print("GITHUB_ACCESS_TOKEN not set in environment variables")
        return

    github_integration = GitHubIntegration(token)

    def test_get_existing_repo():
        repo_name = "isavita/sudoku"
        repo = github_integration.get_repo(repo_name)
        if repo and repo.full_name == repo_name:
            print(f"✅ Successfully fetched repository: {repo.full_name}")
        else:
            print(f"❌ Failed to fetch existing repository: {repo_name}")
        return repo

    def test_get_non_existent_repo():
        repo_name = "isavita/doesnotexist"
        repo = github_integration.get_repo(repo_name)
        if repo is None:
            print(f"✅ Correctly returned None for non-existent repository: {repo_name}")
        else:
            print(f"❌ Unexpectedly fetched a repository for: {repo_name}")

    def test_get_pull_requests(repo):
        if repo is None:
            print("❌ Cannot test pull requests: repository not found")
            return

        open_prs = github_integration.get_pull_requests(repo, state='open')
        print(f"Found {len(open_prs)} open pull requests")

        if open_prs:
            # Test existing PR
            pr_number = open_prs[0].number
            pr = github_integration.get_pull_request(repo, pr_number)
            if pr and pr.number == pr_number:
                print(f"✅ Successfully fetched existing PR #{pr_number}")
            else:
                print(f"❌ Failed to fetch existing PR #{pr_number}")

        # Test non-existent PR
        non_existent_pr = 404  # Using 404 as a likely non-existent PR number
        pr = github_integration.get_pull_request(repo, non_existent_pr)
        if pr is None:
            print(f"✅ Correctly returned None for non-existent PR #{non_existent_pr}")
        else:
            print(f"❌ Unexpectedly fetched a PR for non-existent number: {non_existent_pr}")

        return open_prs

    def test_get_pr_files_and_patches(open_prs):
        if not open_prs:
            print("❌ Cannot test PR files and patches: no open pull requests found")
            return

        # Test PR with files
        pr_with_files = next((pr for pr in open_prs if pr.changed_files > 0), None)
        if pr_with_files:
            files = github_integration.get_pr_files(pr_with_files)
            if files:
                print(f"✅ Successfully fetched {len(files)} files for PR #{pr_with_files.number}")
                
                # Test file patch
                file_with_patch = next((file for file in files if file.patch), None)
                if file_with_patch:
                    patch = github_integration.get_file_patch(file_with_patch)
                    if patch:
                        print(f"✅ Successfully fetched patch for file: {file_with_patch.filename}")
                        print(f"   Patch preview: {patch[:100]}...")  # Print first 100 characters of the patch
                    else:
                        print(f"❌ Failed to fetch patch for file: {file_with_patch.filename}")
                else:
                    print("ℹ️ No files with patches found in this PR")

                # Test file without patch (if exists)
                file_without_patch = next((file for file in files if not file.patch), None)
                if file_without_patch:
                    patch = github_integration.get_file_patch(file_without_patch)
                    if patch is None or patch == "":
                        print(f"✅ Correctly handled file without patch: {file_without_patch.filename}")
                    else:
                        print(f"❌ Unexpectedly fetched patch for file without changes: {file_without_patch.filename}")
                else:
                    print("ℹ️ No files without patches found in this PR")
            else:
                print(f"❌ Failed to fetch files for PR #{pr_with_files.number}")
        else:
            print("ℹ️ No PRs with changed files found to test")

    def test_parse_patch():
        sample_patch = """@@ -18,3 +18,6 @@ This is a simple Sudoku puzzle game implemented in Python using Pygame.
 
 ## Usage
 Run the game by executing:
+```
+python sudoku_game.py
+```"""

        parsed_patch = github_integration.parse_patch(sample_patch)
        
        # Verify parsed information
        assert parsed_patch['old_start'] == 18, f"Old start line should be 18, got {parsed_patch['old_start']}"
        assert parsed_patch['old_lines'] == 3, f"Old line count should be 3, got {parsed_patch['old_lines']}"
        assert parsed_patch['new_start'] == 18, f"New start line should be 18, got {parsed_patch['new_start']}"
        assert parsed_patch['new_lines'] == 6, f"New line count should be 6, got {parsed_patch['new_lines']}"
        assert len(parsed_patch['additions']) == 3, f"Should have 3 additions, got {len(parsed_patch['additions'])}"
        assert len(parsed_patch['deletions']) == 0, f"Should have 0 deletions, got {len(parsed_patch['deletions'])}"
        assert len(parsed_patch['context']) == 3, f"Should have 3 context lines, got {len(parsed_patch['context'])}"
        
        print("✅ Successfully parsed patch")
        print(f"   Additions: {parsed_patch['additions']}")
        print(f"   Context: {parsed_patch['context']}")

    def test_create_and_fetch_comment(repo):
        if repo is None:
            print("❌ Cannot test comment creation: repository not found")
            return
    
        open_prs = github_integration.get_pull_requests(repo, state='open')
        if not open_prs:
            print("❌ Cannot test comment creation: no open pull requests found")
            return
    
        pr = open_prs[0]
        files = github_integration.get_pr_files(pr)
        if not files:
            print("❌ Cannot test comment creation: no files found in the pull request")
            return
    
        file = files[0]
        commit_sha = pr.head.sha
        commit = repo.get_commit(sha=commit_sha)
        body = "This is a test comment created by the GitHub Integration test suite."
    
        # Parse the patch to find a valid position
        patch_data = github_integration.parse_patch(file.patch)
        if not patch_data:
            print("❌ Cannot test comment creation: failed to parse patch")
            return
    
        # Use the end_line from patch_data as the position
        position = patch_data['end_line']
    
        try:
            print(f"commit: {commit}, type: {type(commit)}")
            print(f"path: {file.filename}, type: {type(file.filename)}")
            print(f"position: {position}, type: {type(position)}")
            print(f"Patch data: {patch_data}")
    
            comment = github_integration.create_comment(pr, body, commit, file.filename, position)
            print(f"✅ Successfully created comment on PR #{pr.number}")
            print(f"   Comment ID: {comment.id}")
            print(f"   Comment body: {comment.body}")
            print(f"   Comment position: {comment.position}")
            print(f"   File: {file.filename}")
            print(f"   Commit SHA: {commit_sha}")
    
            # Verify we can fetch the created comment
            fetched_comment = pr.get_review_comment(comment.id)
            if fetched_comment.body == body:
                print("✅ Successfully fetched the created comment")
            else:
                print("❌ Fetched comment does not match the created comment")
    
            # Clean up: delete the test comment
            comment.delete()
            print("✅ Test comment deleted")
    
        except Exception as e:
            print(f"❌ Failed to create or fetch comment: {str(e)}")
            print(f"   Error type: {type(e)}")
            import traceback
            traceback.print_exc()
            if hasattr(e, 'data'):
                print(f"   Error data: {e.data}")
    
    
    print("Running tests...")
    repo = test_get_existing_repo()
    test_get_non_existent_repo()
    open_prs = test_get_pull_requests(repo)
    test_get_pr_files_and_patches(open_prs)
    test_parse_patch()
    test_create_and_fetch_comment(repo)
    
# Run the tests
# run_tests()
````

````python
import os
import re
from github import Github, GithubException

class GitHubIntegration:
    def __init__(self, access_token):
        self.github = Github(access_token)

    def get_repo(self, repo_name):
        try:
            return self.github.get_repo(repo_name)
        except GithubException as e:
            if e.status == 404:
                return None
            else:
                raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_pull_request(self, repo, pr_number):
        try:
            return repo.get_pull(pr_number)
        except GithubException as e:
            if e.status == 404:
                return None
            else:
                raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_pull_requests(self, repo, state='open'):
        return list(repo.get_pulls(state=state))

    def get_pr_files(self, pull_request):
        try:
            return list(pull_request.get_files())
        except GithubException as e:
            raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_file_patch(self, file):
        return file.patch

    def parse_patch(self, patch):
        if not patch:
            return None  # Return None for empty patches

        lines = patch.split('\n')
        header = lines[0]
        changes = lines[1:]
        
        # Parse the header
        match = re.match(r'@@ -(\d+)(?:,(\d+))? \+(\d+)(?:,(\d+))? @@', header)
        if not match:
            raise ValueError("Invalid patch header format")
        
        old_start = int(match.group(1))
        old_lines = int(match.group(2) or 1)  # Default to 1 if not specified
        new_start = int(match.group(3))
        new_lines = int(match.group(4) or 1)  # Default to 1 if not specified
        
        additions = [line[1:] for line in changes if line.startswith('+')]
        deletions = [line[1:] for line in changes if line.startswith('-')]
        context = [line[1:] for line in changes if line.startswith(' ')]
        
        is_new_file = old_start == 0 and old_lines == 0
        is_deleted_file = new_start == 0 and new_lines == 0
        
        if is_new_file:
            end_line = new_start + len(additions) - 1
        elif is_deleted_file:
            end_line = old_start + len(deletions) - 1
        else:
            end_line = new_start + len(additions) - len(deletions) + old_lines - 1
        
        return {
            'old_start': old_start,
            'old_lines': old_lines,
            'new_start': new_start,
            'new_lines': new_lines,
            'additions': additions,
            'deletions': deletions,
            'context': context,
            'start_position': 1,  # GitHub API uses 1-based indexing for positions
            'end_position': len(changes),
            'end_line': max(1, end_line),  # Ensure end_line is at least 1
            'is_new_file': is_new_file,
            'is_deleted_file': is_deleted_file
        }

    def create_comment(self, pull_request, body, commit, path, position):
        try:
            comment = pull_request.create_review_comment(body, commit, path, position)
            return comment
        except GithubException as e:
            raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")

    def get_commit(self, repo, commit_sha):
        try:
            return repo.get_commit(commit_sha)
        except GithubException as e:
            if e.status in [404, 422]:  # Both 404 and 422 can indicate a non-existent commit
                return None
            else:
                raise RuntimeError(f"GitHub API error: {e.data.get('message', str(e))}")


def run_tests():
    token = os.getenv("GITHUB_ACCESS_TOKEN")
    if not token:
        print("GITHUB_ACCESS_TOKEN not set in environment variables")
        return

    github_integration = GitHubIntegration(token)

    def test_get_existing_repo():
        repo_name = "isavita/sudoku"
        repo = github_integration.get_repo(repo_name)
        if repo and repo.full_name == repo_name:
            print(f"✅ Successfully fetched repository: {repo.full_name}")
        else:
            print(f"❌ Failed to fetch existing repository: {repo_name}")
        return repo

    def test_get_non_existent_repo():
        repo_name = "isavita/doesnotexist"
        repo = github_integration.get_repo(repo_name)
        if repo is None:
            print(f"✅ Correctly returned None for non-existent repository: {repo_name}")
        else:
            print(f"❌ Unexpectedly fetched a repository for: {repo_name}")

    def test_get_pull_requests(repo):
        if repo is None:
            print("❌ Cannot test pull requests: repository not found")
            return

        open_prs = github_integration.get_pull_requests(repo, state='open')
        print(f"Found {len(open_prs)} open pull requests")

        if open_prs:
            # Test existing PR
            pr_number = open_prs[0].number
            pr = github_integration.get_pull_request(repo, pr_number)
            if pr and pr.number == pr_number:
                print(f"✅ Successfully fetched existing PR #{pr_number}")
            else:
                print(f"❌ Failed to fetch existing PR #{pr_number}")

        # Test non-existent PR
        non_existent_pr = 404  # Using 404 as a likely non-existent PR number
        pr = github_integration.get_pull_request(repo, non_existent_pr)
        if pr is None:
            print(f"✅ Correctly returned None for non-existent PR #{non_existent_pr}")
        else:
            print(f"❌ Unexpectedly fetched a PR for non-existent number: {non_existent_pr}")

        return open_prs

    def test_get_pr_files_and_patches(open_prs):
        if not open_prs:
            print("❌ Cannot test PR files and patches: no open pull requests found")
            return

        # Test PR with files
        pr_with_files = next((pr for pr in open_prs if pr.changed_files > 0), None)
        if pr_with_files:
            files = github_integration.get_pr_files(pr_with_files)
            if files:
                print(f"✅ Successfully fetched {len(files)} files for PR #{pr_with_files.number}")
                
                # Test file patch
                file_with_patch = next((file for file in files if file.patch), None)
                if file_with_patch:
                    patch = github_integration.get_file_patch(file_with_patch)
                    if patch:
                        print(f"✅ Successfully fetched patch for file: {file_with_patch.filename}")
                        print(f"   Patch preview: {patch[:100]}...")  # Print first 100 characters of the patch
                    else:
                        print(f"❌ Failed to fetch patch for file: {file_with_patch.filename}")
                else:
                    print("ℹ️ No files with patches found in this PR")

                # Test file without patch (if exists)
                file_without_patch = next((file for file in files if not file.patch), None)
                if file_without_patch:
                    patch = github_integration.get_file_patch(file_without_patch)
                    if patch is None or patch == "":
                        print(f"✅ Correctly handled file without patch: {file_without_patch.filename}")
                    else:
                        print(f"❌ Unexpectedly fetched patch for file without changes: {file_without_patch.filename}")
                else:
                    print("ℹ️ No files without patches found in this PR")
            else:
                print(f"❌ Failed to fetch files for PR #{pr_with_files.number}")
        else:
            print("ℹ️ No PRs with changed files found to test")

    def test_parse_patch():
        sample_patch = """@@ -18,3 +18,6 @@ This is a simple Sudoku puzzle game implemented in Python using Pygame.
 
 ## Usage
 Run the game by executing:
+```
+python sudoku_game.py
+```"""

        parsed_patch = github_integration.parse_patch(sample_patch)
        
        # Verify parsed information
        assert parsed_patch['old_start'] == 18, f"Old start line should be 18, got {parsed_patch['old_start']}"
        assert parsed_patch['old_lines'] == 3, f"Old line count should be 3, got {parsed_patch['old_lines']}"
        assert parsed_patch['new_start'] == 18, f"New start line should be 18, got {parsed_patch['new_start']}"
        assert parsed_patch['new_lines'] == 6, f"New line count should be 6, got {parsed_patch['new_lines']}"
        assert len(parsed_patch['additions']) == 3, f"Should have 3 additions, got {len(parsed_patch['additions'])}"
        assert len(parsed_patch['deletions']) == 0, f"Should have 0 deletions, got {len(parsed_patch['deletions'])}"
        assert len(parsed_patch['context']) == 3, f"Should have 3 context lines, got {len(parsed_patch['context'])}"
        
        print("✅ Successfully parsed patch")
        print(f"   Additions: {parsed_patch['additions']}")
        print(f"   Context: {parsed_patch['context']}")

    def test_create_and_fetch_comment(repo):
        if repo is None:
            print("❌ Cannot test comment creation: repository not found")
            return
    
        open_prs = github_integration.get_pull_requests(repo, state='open')
        if not open_prs:
            print("❌ Cannot test comment creation: no open pull requests found")
            return
    
        pr = open_prs[0]
        files = github_integration.get_pr_files(pr)
        if not files:
            print("❌ Cannot test comment creation: no files found in the pull request")
            return
    
        file = files[0]
        commit_sha = pr.head.sha
        commit = repo.get_commit(sha=commit_sha)
        body = "This is a test comment created by the GitHub Integration test suite."
    
        # Parse the patch to find a valid position
        patch_data = github_integration.parse_patch(file.patch)
        if not patch_data:
            print("❌ Cannot test comment creation: failed to parse patch")
            return
    
        # Use the end_line from patch_data as the position
        position = patch_data['end_line']
    
        try:
            print(f"commit: {commit}, type: {type(commit)}")
            print(f"path: {file.filename}, type: {type(file.filename)}")
            print(f"position: {position}, type: {type(position)}")
            print(f"Patch data: {patch_data}")
    
            comment = github_integration.create_comment(pr, body, commit, file.filename, position)
            print(f"✅ Successfully created comment on PR #{pr.number}")
            print(f"   Comment ID: {comment.id}")
            print(f"   Comment body: {comment.body}")
            print(f"   Comment position: {comment.position}")
            print(f"   File: {file.filename}")
            print(f"   Commit SHA: {commit_sha}")
    
            # Verify we can fetch the created comment
            fetched_comment = pr.get_review_comment(comment.id)
            if fetched_comment.body == body:
                print("✅ Successfully fetched the created comment")
            else:
                print("❌ Fetched comment does not match the created comment")
    
            # Clean up: delete the test comment
            comment.delete()
            print("✅ Test comment deleted")
    
        except Exception as e:
            print(f"❌ Failed to create or fetch comment: {str(e)}")
            print(f"   Error type: {type(e)}")
            import traceback
            traceback.print_exc()
            if hasattr(e, 'data'):
                print(f"   Error data: {e.data}")
        
    def test_get_commit(repo):
        if repo is None:
            print("❌ Cannot test get_commit: repository not found")
            return

        # Get the latest commit SHA
        latest_commit = repo.get_commits()[0]
        latest_commit_sha = latest_commit.sha

        # Test getting an existing commit
        commit = github_integration.get_commit(repo, latest_commit_sha)
        if commit and commit.sha == latest_commit_sha:
            print(f"✅ Successfully fetched existing commit: {latest_commit_sha[:7]}")
            print(f"   Commit message: {commit.commit.message[:50]}...")  # Print first 50 chars of commit message
        else:
            print(f"❌ Failed to fetch existing commit: {latest_commit_sha[:7]}")

        # Test getting a non-existent commit
        non_existent_sha = "1" * 40  # A SHA that's unlikely to exist
        non_existent_commit = github_integration.get_commit(repo, non_existent_sha)
        if non_existent_commit is None:
            print(f"✅ Correctly returned None for non-existent commit: {non_existent_sha[:7]}")
        else:
            print(f"❌ Unexpectedly fetched a commit for non-existent SHA: {non_existent_sha[:7]}")

    print("Running tests...")
    repo = test_get_existing_repo()
    test_get_non_existent_repo()
    open_prs = test_get_pull_requests(repo)
    test_get_pr_files_and_patches(open_prs)
    test_parse_patch()
    test_create_and_fetch_comment(repo)
    test_get_commit(repo)
    
# Run the tests
# run_tests()
````
